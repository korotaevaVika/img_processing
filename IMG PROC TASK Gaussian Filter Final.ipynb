```python
%matplotlib inline 
   
import ipywidgets as widgets    
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
from IPython.display import clear_output
```

```python
source_img_rgb = cv2.imread('original.png')
print("Source Image (RGB)")
plt.imshow(source_img_rgb) # Source Image (RGB)
plt.show()
```

```python
param = widgets.IntSlider(
    value=5,
    min=1,
    max=11,
    step=2,
    description='Kernel Size Param:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
display(param)
print(param.value)

def on_value_change(change):
    i = int(param.value)
    display(param)
    try:
        # GaussianBlur( кадр, размер ядра, отклонение)
        img = cv2.GaussianBlur(source_img_rgb, (i, i), 0)
        clear_output()
        display(param)
        print('Kernel Size Param = ', i)
        plt.imshow(img)
        plt.show()
    except:
        clear_output()
        display(param)
        print('Kernel Size Param = ', i)
        print('Exception catched')
        

param.observe(on_value_change, names='value')
```

```python
source_images = [
    cv2.imread('original.png'), 
    cv2.imread('original.png')[10:60, 10:60], 
    cv2.imread('original.png'), 
    cv2.imread('original.png')
]

min_w = source_images[0].shape[0]
min_h = source_images[0].shape[1]
for val in source_images:
    if (val.shape[0] < min_w):
        min_w = val.shape[0]
    if (val.shape[1] < min_h):
        min_h = val.shape[1]


for i in np.arange(len(source_images)):
    source_images[i] = source_images[i][0:min_w, 0:min_h]

index = 0
i = 5
#process 1st image
source_images[index] = cv2.GaussianBlur(source_images[index], (i, i), 0)
tlt = "Image "+ str(index) + " Kernel_Size = (" + str(i) + ", " + str(i) + ") "
plt.subplot(221),plt.imshow(source_images[index]),plt.title(tlt)
plt.xticks([]), plt.yticks([])

index = 1
i = 3
#process 2nd image
source_images[index] = cv2.GaussianBlur(source_images[index], (i, i), 0)
tlt = "Image "+ str(index) + " Kernel_Size = (" + str(i) + ", " + str(i) + ") "
plt.subplot(222),plt.imshow(source_images[index]),plt.title(tlt)
plt.xticks([]), plt.yticks([])

index = 2
i = 7
#process 3d image
source_images[index] = cv2.GaussianBlur(source_images[index], (i, i), 0)
tlt = "Image "+ str(index) + " Kernel_Size = (" + str(i) + ", " + str(i) + ") "
plt.subplot(223),plt.imshow(source_images[index]),plt.title(tlt)
plt.xticks([]), plt.yticks([])

index = 3
i = 9
#process 4th image
source_images[index] = cv2.GaussianBlur(source_images[index], (i, i), 0)
tlt = "Image "+ str(index) + " Kernel_Size = (" + str(i) + ", " + str(i) + ") "
plt.subplot(224),plt.imshow(source_images[index]),plt.title(tlt)
plt.xticks([]), plt.yticks([])

plt.show()
```
